7.DATA QUALITY ANALYSIS

A. Missing Values
The dataset contains missing values in the following columns:
• Description: 1,454 missing values
• CustomerID: 135,080 missing values
Missing CustomerID values significantly impact customer-level analysis such as RFM modeling and clustering. Therefore, records without CustomerID will be removed during preprocessing.

B. Abnormal and Inconsistent Values
The dataset contains abnormal values that require cleaning:
• Quantity ≤ 0: 10,624 records (returns or cancellations)
• UnitPrice ≤ 0: 2,517 records
Negative quantities likely represent returned transactions, while zero or negative unit prices may indicate data entry issues. These records will be filtered out for modeling tasks.

C. Duplicate Records
Preliminary inspection suggests that some duplicate invoice- product combinations exist. Duplicate entries will be checked and handled to avoid bias in association rule mining.
